```python
# ============================================================
# STEP 0: Import Libraries
# ============================================================

import os
import sys
import shutil
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import ResNet50V2, EfficientNetV2B0
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    GlobalAveragePooling2D, Dense, Dropout, BatchNormalization,
    Input, Concatenate, Lambda
)
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import (
    ReduceLROnPlateau, EarlyStopping, ModelCheckpoint, LearningRateScheduler
)
from tensorflow.keras import regularizers


# ============================================================
# STEP 1: Complete Error Suppression
# ============================================================

os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
os.environ['TF_ENABLE_ONEDNN_OPTS'] = '0'
os.environ['CUDA_VISIBLE_DEVICES'] = '0'

import logging
logging.getLogger('tensorflow').setLevel(logging.ERROR)

import warnings
warnings.filterwarnings('ignore')

stderr = sys.stderr
sys.stderr = open(os.devnull, 'w')

try:
    gpus = tf.config.list_physical_devices('GPU')
    if gpus:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        gpu_available = True
    else:
        gpu_available = False
except:
    gpu_available = False

sys.stderr = stderr

if gpu_available:
    print("✅ GPU Enabled: Training accelerated!")
else:
    print("⚙️ Running on CPU")


# ============================================================
# STEP 2: Find All Datasets
# ============================================================

def find_all_dataset_roots(base="/kaggle/input", min_class_folders=2):
    dataset_paths = []
    visited = set()

    for root, dirs, files in os.walk(base):
        if root in visited:
            continue
            
        count = 0
        for d in dirs:
            full_path = os.path.join(root, d)
            try:
                has_images = any(
                    f.lower().endswith(('.jpg', '.jpeg', '.png'))
                    for f in os.listdir(full_path)
                    if os.path.isfile(os.path.join(full_path, f))
                )
                if has_images:
                    count += 1
            except:
                continue

        if count >= min_class_folders:
            dataset_paths.append(root)
            visited.add(root)

    return dataset_paths


all_datasets = find_all_dataset_roots()

if not all_datasets:
    raise FileNotFoundError("❌ No datasets found")

print(f"\n🔍 Found {len(all_datasets)} dataset(s)")


# ============================================================
# STEP 3: Setup Working Directories
# ============================================================

base_dir  = "/kaggle/working/brain_tumor_split"
train_dir = os.path.join(base_dir, "train")
val_dir   = os.path.join(base_dir, "val")
test_dir  = os.path.join(base_dir, "test")

if os.path.exists(base_dir):
    shutil.rmtree(base_dir)

os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)


# ============================================================
# STEP 4: Merge Multiple Datasets
# ============================================================

standard_classes = ['glioma_tumor', 'meningioma_tumor', 'no_tumor', 'pituitary_tumor']

class_mapping = {
    'glioma': 'glioma_tumor',
    'meningioma': 'meningioma_tumor',
    'notumor': 'no_tumor',
    'pituitary': 'pituitary_tumor',
    'glioma_tumor': 'glioma_tumor',
    'meningioma_tumor': 'meningioma_tumor',
    'no_tumor': 'no_tumor',
    'pituitary_tumor': 'pituitary_tumor'
}

print(f"\n🧠 Merging datasets into 4 classes\n")

collected_images = {cls: [] for cls in standard_classes}

for dataset_path in all_datasets:
    try:
        all_classes = [
            d for d in os.listdir(dataset_path)
            if os.path.isdir(os.path.join(dataset_path, d))
        ]
    except:
        continue
    
    for class_name in all_classes:
        class_path = os.path.join(dataset_path, class_name)
        normalized_class = class_mapping.get(class_name.lower())
        if not normalized_class:
            continue
        
        try:
            images = [
                os.path.join(class_path, f) for f in os.listdir(class_path)
                if f.lower().endswith(('.png', '.jpg', '.jpeg'))
            ]
            collected_images[normalized_class].extend(images)
        except:
            continue

print(f"📊 Total images per class:")
total_images = 0
for cls, imgs in collected_images.items():
    print(f"   {cls:<25} : {len(imgs)} images")
    total_images += len(imgs)
print(f"   {'TOTAL':<25} : {total_images} images")


# ============================================================
# STEP 5: Train-Val-Test Split (70-15-15)
# ============================================================

print(f"\n✂️ Splitting into train/val/test (70/15/15)...\n")

for class_name, image_paths in collected_images.items():
    if not image_paths:
        continue
    
    # Use ALL available images (no sampling)
    np.random.shuffle(image_paths)
    
    # 70% train, 15% val, 15% test
    train_split = int(0.70 * len(image_paths))
    val_split = int(0.85 * len(image_paths))
    
    train_files = image_paths[:train_split]
    val_files = image_paths[train_split:val_split]
    test_files = image_paths[val_split:]

    os.makedirs(os.path.join(train_dir, class_name), exist_ok=True)
    os.makedirs(os.path.join(val_dir, class_name), exist_ok=True)
    os.makedirs(os.path.join(test_dir, class_name), exist_ok=True)

    for src_path in train_files:
        try:
            shutil.copy(src_path, os.path.join(train_dir, class_name, os.path.basename(src_path)))
        except:
            continue

    for src_path in val_files:
        try:
            shutil.copy(src_path, os.path.join(val_dir, class_name, os.path.basename(src_path)))
        except:
            continue

    for src_path in test_files:
        try:
            shutil.copy(src_path, os.path.join(test_dir, class_name, os.path.basename(src_path)))
        except:
            continue

    print(f"  ✅ {class_name:<20} Train: {len(train_files):4d} | Val: {len(val_files):4d} | Test: {len(test_files):4d}")

print("\n✅ Split completed!")


# ============================================================
# STEP 6: MEDICAL-GRADE Data Augmentation
# ============================================================

print("\n🔬 Applying medical-grade augmentation...\n")

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    zoom_range=0.25,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.15,
    horizontal_flip=True,
    vertical_flip=True,
    brightness_range=[0.6, 1.4],
    fill_mode='reflect',
    channel_shift_range=20.0
)

val_test_datagen = ImageDataGenerator(rescale=1./255)

# Use 256x256 for better feature extraction
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(256, 256),
    batch_size=16,
    class_mode="categorical",
    shuffle=True
)

val_generator = val_test_datagen.flow_from_directory(
    val_dir,
    target_size=(256, 256),
    batch_size=16,
    class_mode="categorical",
    shuffle=False
)

test_generator = val_test_datagen.flow_from_directory(
    test_dir,
    target_size=(256, 256),
    batch_size=16,
    class_mode="categorical",
    shuffle=False
)


# ============================================================
# STEP 7: Show Data Info
# ============================================================

print("📊 Final Dataset Info:")
print(f"   Classes      : {list(train_generator.class_indices.keys())}")
print(f"   Train samples: {train_generator.samples}")
print(f"   Val samples  : {val_generator.samples}")
print(f"   Test samples : {test_generator.samples}\n")


# ============================================================
# STEP 8: Build Advanced Ensemble Model
# ============================================================

print("🏗️ Building Advanced Ensemble Architecture...\n")

input_layer = Input(shape=(256, 256, 3))

# Branch 1: EfficientNetV2B0
base_model_1 = EfficientNetV2B0(
    weights='imagenet',
    include_top=False,
    input_tensor=input_layer
)
base_model_1.trainable = False
x1 = GlobalAveragePooling2D()(base_model_1.output)

# Branch 2: ResNet50V2
base_model_2 = ResNet50V2(
    weights='imagenet',
    include_top=False,
    input_tensor=input_layer
)
base_model_2.trainable = False
x2 = GlobalAveragePooling2D()(base_model_2.output)

# Concatenate both branches
combined = Concatenate()([x1, x2])

# Classification head with heavy regularization
x = BatchNormalization()(combined)
x = Dense(1024, activation='relu', kernel_regularizer=regularizers.l2(0.001))(x)
x = Dropout(0.6)(x)
x = BatchNormalization()(x)
x = Dense(512, activation='relu', kernel_regularizer=regularizers.l2(0.001))(x)
x = Dropout(0.5)(x)
x = BatchNormalization()(x)
x = Dense(256, activation='relu', kernel_regularizer=regularizers.l2(0.001))(x)
x = Dropout(0.4)(x)
output = Dense(4, activation='softmax')(x)

model = Model(inputs=input_layer, outputs=output)

print(f"Total Parameters: {model.count_params():,}")
print(f"Trainable Parameters: {sum([tf.size(w).numpy() for w in model.trainable_weights]):,}\n")


# ============================================================
# STEP 9: Advanced Training Configuration
# ============================================================

# Cosine annealing learning rate schedule
def cosine_annealing(epoch, lr):
    epochs = 25
    return 0.001 * (1 + np.cos(np.pi * epoch / epochs)) / 2

model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

checkpoint = ModelCheckpoint(
    '/kaggle/working/best_model.keras',
    monitor='val_accuracy',
    save_best_only=True,
    mode='max',
    verbose=0
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_accuracy',
    factor=0.3,
    patience=4,
    min_lr=1e-7,
    mode='max',
    verbose=1
)

early_stop = EarlyStopping(
    monitor='val_accuracy',
    patience=12,
    restore_best_weights=True,
    mode='max',
    verbose=1
)

lr_scheduler = LearningRateScheduler(cosine_annealing, verbose=0)


# ============================================================
# STEP 10: Phase 1 - Train Classifier Head
# ============================================================

print("🚀 Phase 1: Training classifier head (frozen base)...\n")

history1 = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=20,
    callbacks=[checkpoint, reduce_lr, early_stop],
    verbose=1
)


# ============================================================
# STEP 11: Phase 2 - Fine-tune Entire Model
# ============================================================

print("\n🔥 Phase 2: Fine-tuning entire model...\n")

# Unfreeze all layers
base_model_1.trainable = True
base_model_2.trainable = True

# Freeze batch norm layers for stability
for layer in base_model_1.layers:
    if isinstance(layer, BatchNormalization):
        layer.trainable = False

for layer in base_model_2.layers:
    if isinstance(layer, BatchNormalization):
        layer.trainable = False

model.compile(
    optimizer=Adam(learning_rate=1e-5),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

history2 = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=20,
    callbacks=[checkpoint, reduce_lr, early_stop],
    verbose=1
)


# ============================================================
# STEP 12: Evaluate on Test Set
# ============================================================

# Load best weights
model.load_weights('/kaggle/working/best_model.keras')

loss, accuracy = model.evaluate(test_generator, verbose=0)
print(f"\n{'='*60}")
print(f"🏆 FINAL TEST ACCURACY : {accuracy * 100:.2f}%")
print(f"📉 FINAL TEST LOSS     : {loss:.4f}")
print(f"{'='*60}\n")


# ============================================================
# STEP 13: Training Graphs
# ============================================================

combined_acc = history1.history['accuracy'] + history2.history['accuracy']
combined_val_acc = history1.history['val_accuracy'] + history2.history['val_accuracy']
combined_loss = history1.history['loss'] + history2.history['loss']
combined_val_loss = history1.history['val_loss'] + history2.history['val_loss']

plt.figure(figsize=(16, 6))

plt.subplot(1, 2, 1)
plt.plot(combined_acc, label="Train Accuracy", linewidth=2.5)
plt.plot(combined_val_acc, label="Val Accuracy", linewidth=2.5)
plt.axvline(x=len(history1.history['accuracy']), color='red', linestyle='--', label='Fine-tuning starts', linewidth=2)
plt.legend(fontsize=11)
plt.title("Ensemble Model - Accuracy Progression", fontsize=14, fontweight='bold')
plt.xlabel("Epoch", fontsize=12)
plt.ylabel("Accuracy", fontsize=12)
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(combined_loss, label="Train Loss", linewidth=2.5)
plt.plot(combined_val_loss, label="Val Loss", linewidth=2.5)
plt.axvline(x=len(history1.history['loss']), color='red', linestyle='--', label='Fine-tuning starts', linewidth=2)
plt.legend(fontsize=11)
plt.title("Ensemble Model - Loss Progression", fontsize=14, fontweight='bold')
plt.xlabel("Epoch", fontsize=12)
plt.ylabel("Loss", fontsize=12)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()


# ============================================================
# STEP 14: Detailed Classification Report
# ============================================================

print("\n🔍 Generating predictions...")
pred_probs = model.predict(test_generator, verbose=0)
y_pred = np.argmax(pred_probs, axis=1)
class_names = list(train_generator.class_indices.keys())

print("\n📊 CLASSIFICATION REPORT:")
print("="*80)
print(classification_report(test_generator.classes, y_pred, target_names=class_names, digits=4))

print("\n📊 CONFUSION MATRIX:")
cm = confusion_matrix(test_generator.classes, y_pred)
print(cm)

# Confusion Matrix Visualization
plt.figure(figsize=(10, 8))
plt.imshow(cm, interpolation='nearest', cmap='YlGnBu')
plt.title('Confusion Matrix - Brain Tumor Classification', fontsize=14, fontweight='bold', pad=20)
plt.colorbar()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names, rotation=45, ha='right', fontsize=11)
plt.yticks(tick_marks, class_names, fontsize=11)

for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, str(cm[i, j]), ha="center", va="center", 
                color="white" if cm[i, j] > cm.max() / 2 else "black",
                fontsize=13, fontweight='bold')

plt.ylabel('True Label', fontsize=12, fontweight='bold')
plt.xlabel('Predicted Label', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()


# ============================================================
# STEP 15: Per-Class Accuracy
# ============================================================

print("\n📈 PER-CLASS ACCURACY:")
print("="*80)
for i, class_name in enumerate(class_names):
    class_mask = test_generator.classes == i
    class_acc = np.mean(y_pred[class_mask] == i) * 100
    print(f"   {class_name:<20} : {class_acc:6.2f}%")
print("="*80)


# ============================================================
# STEP 16: Visual Predictions with Confidence
# ============================================================

test_batch = next(test_generator)
test_images = test_batch[0]
test_labels = test_batch[1]

print("\n🔬 SAMPLE PREDICTIONS:\n")

plt.figure(figsize=(18, 12))

for i in range(min(12, len(test_images))):
    img = test_images[i]
    true_label = class_names[np.argmax(test_labels[i])]
    
    arr = np.expand_dims(img, axis=0)
    pred_prob = model.predict(arr, verbose=0)
    pred = np.argmax(pred_prob)
    confidence = pred_prob[0][pred] * 100
    predicted_label = class_names[pred]
    
    is_correct = predicted_label == true_label
    title_color = "green" if is_correct else "red"
    
    print(f"Sample {i+1:2d}: True={true_label:<20} | Pred={predicted_label:<20} | Conf={confidence:5.1f}% | {'✅' if is_correct else '❌'}")
    
    plt.subplot(3, 4, i+1)
    plt.imshow(img)
    plt.title(f"Pred: {predicted_label}\nConf: {confidence:.1f}%\nTrue: {true_label}", 
             color=title_color, fontsize=10, fontweight='bold')
    plt.axis("off")

plt.tight_layout()
plt.show()

print(f"\n{'='*80}")
print(f"🎓 ACADEMIC PROJECT COMPLETE!")
print(f"🏆 Final Accuracy: {accuracy * 100:.2f}%")
print(f"{'='*80}\n")
```
